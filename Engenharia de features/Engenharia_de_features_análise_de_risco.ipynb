```python
import pandas as pd

df = pd.read_csv('/content/df_final_selected_features.csv')
df
```

```python
df.columns
```

```python
''' Feature: idade em anos
Importância: captura o efeito gradual da maturidade financeira no risco '''

# Transforma dias de nascimento em idade(anos)
df['AGE_YEARS'] = (-df['DAYS_BIRTH'] / 365).astype(int)
```

```python
''' Feature: faixas etárias baseadas no EDA
Importância: modela a relação não linear entre idade e inadimplência '''

# Criar faixas etárias a partir da idade em anos
df['AGE_BIN'] = pd.cut(
    df['AGE_YEARS'],                # Coluna com a idade do cliente
    bins=[18, 25, 35, 45, 55, 100],  # Limites das faixas etárias
    labels=['18-25', '26-35', '36-45', '46-55', '56+']  # Rótulos das faixas
)
```

```python
''' Feature: indicador de cliente jovem
Importância: clientes jovens concentraram maior risco no EDA '''

# Criar flag para identificar clientes jovens
df['IS_YOUNG_CLIENT'] = (df['AGE_YEARS'] < 30).astype(int)  # 1 se idade < 30, senão 0
```

```python
''' Feature: indicador de cliente mais velho
Importância: clientes mais velhos apresentaram menor inadimplência '''

# Criar flag para identificar clientes mais velhos
df['IS_SENIOR_CLIENT'] = (df['AGE_YEARS'] > 55).astype(int)
```

```python
import numpy as np

''' Feature: dívida ativa total em escala log
Importância: reduz assimetria e impacto de valores extremos '''

# Aplicar transformação logarítmica para reduzir assimetria da dívida ativa
df['DEBT_ACTIVE_LOG'] = np.log1p(df['BUREAU_AMT_CREDIT_SUM_DEBT_SUM'])  # log(1 + dívida)
```

```python
''' Feature: faixas de endividamento por quantis
Importância: captura diferentes níveis relativos de risco de dívida '''

# Criar faixas de dívida ativa com base em quantis
df['DEBT_BIN'] = pd.qcut(
    df['BUREAU_AMT_CREDIT_SUM_DEBT_SUM'],
    q=5,
    duplicates='drop'
)
```

```python
''' Feature: flag de alto endividamento
Importância: identifica clientes concentrados nas caudas de risco '''

# Calcular o percentil 80 da dívida ativa
debt_threshold = df['BUREAU_AMT_CREDIT_SUM_DEBT_SUM'].quantile(0.80)

# Criar flag para clientes com dívida acima do percentil 80
df['HIGH_DEBT_FLAG'] = (df['BUREAU_AMT_CREDIT_SUM_DEBT_SUM'] > debt_threshold).astype(int)
```

```python
''' Feature: interação entre juventude e alto endividamento
Importância: jovens endividados apresentaram risco significativamente maior '''

# Criar flag para clientes jovens com alta dívida
df['YOUNG_HIGH_DEBT'] = (
    (df['IS_YOUNG_CLIENT'] == 1) &   # Cliente jovem
    (df['HIGH_DEBT_FLAG'] == 1)      # Dívida acima do percentil 80
).astype(int)                        # 1 se ambas condições forem verdadeiras
```

```python
''' Feature: interação entre senioridade e baixo endividamento
Importância: identifica clientes com perfil historicamente mais seguro '''

# Criar flag para clientes mais velhos com baixa dívida
df['SENIOR_LOW_DEBT'] = (
    (df['IS_SENIOR_CLIENT'] == 1) &
    (df['HIGH_DEBT_FLAG'] == 0)
).astype(int)
```

```python
''' Feature: renda mensal em escala log
Importância: reduz assimetria e efeito de outliers na renda '''

# # Aplicar transformação logarítmica para reduzir assimetria da renda
df['INCOME_LOG'] = np.log1p(df['AMT_INCOME_TOTAL'])  # log(1 + renda total)
```

```python
''' Feature: faixas de renda por quantis
Importância: captura níveis relativos de capacidade financeira '''

# Criar faixas de renda com base em quantis
df['INCOME_BIN'] = pd.qcut(
    df['AMT_INCOME_TOTAL'],  # Renda total do cliente
    q=5,                     # Divide em 5 grupos com tamanhos similares
    duplicates='drop'
)
```

```python
''' Feature: razão entre valor do crédito e renda
Importância: mede o peso do empréstimo em relação à renda do cliente '''

# # Calcular razão entre valor do crédito e renda do cliente
df['CREDIT_TO_INCOME'] = (
    df['AMT_CREDIT'] / (df['AMT_INCOME_TOTAL'] + 1)  # +1 evita divisão por zero
)
```

```python
''' Feature: razão entre parcela mensal e renda
Importância: aproxima o esforço mensal de pagamento do cliente '''

# Calcular razão entre valor da anuidade e a renda do cliente
df['ANNUITY_TO_INCOME'] = (
    df['AMT_ANNUITY'] / (df['AMT_INCOME_TOTAL'] + 1)  # +1 evita divisão por zero
)
```

```python
''' Feature: indicador de alto comprometimento de renda
Importância: clientes com alta pressão financeira têm maior risco '''

# Criar flag para identificar alta carga de pagamento sobre a renda
df['HIGH_PAYMENT_BURDEN'] = (
    df['ANNUITY_TO_INCOME'] > 0.30  # Parcela anual acima de 30% da renda
).astype(int)                      # 1 = alto comprometimento de renda
```

```python
''' Feature: renda residual após parcela
Importância: estima quanto sobra para despesas básicas '''

# Calcular renda residual após o pagamento da anuidade
df['RESIDUAL_INCOME'] = (
    df['AMT_INCOME_TOTAL'] - df['AMT_ANNUITY']  # Renda disponível após parcela
)
```

```python
''' Feature: renda residual em escala log
Importância: suaviza extremos e melhora estabilidade do modelo '''

# Aplicar log na renda residual para reduzir assimetria e evitar valores negativos
df['RESIDUAL_INCOME_LOG'] = np.log1p(
    df['RESIDUAL_INCOME'].clip(lower=0)  # Trunca valores negativos em zero
)
```

```python
''' Feature: interação entre juventude e alto comprometimento
Importância: jovens com alta pressão financeira têm risco elevado '''

# Criar flag para clientes jovens com alta carga de pagamento
df['YOUNG_HIGH_BURDEN'] = (
    (df['IS_YOUNG_CLIENT'] == 1) &     # Cliente jovem
    (df['HIGH_PAYMENT_BURDEN'] == 1)  # Parcela compromete >30% da renda
).astype(int)                         # 1 se ambas condições forem verdadeiras
```

```python
''' Feature: interação entre senioridade e baixo comprometimento
Importância: perfil financeiramente mais estável '''

# Criar flag para clientes seniores com baixa carga de pagamento
df['SENIOR_LOW_BURDEN'] = (
    (df['IS_SENIOR_CLIENT'] == 1) &     # Cliente sênior
    (df['HIGH_PAYMENT_BURDEN'] == 0)   # Parcela NÃO compromete >30% da renda
).astype(int)                          # 1 se ambas condições forem verdadeiras
```

```python
''' Feature: média de dias desde abertura dos créditos no bureau
Importância: histórico mais antigo indica maior experiência de crédito '''

# Converter dias de crédito no bureau em valor positivo (idade média do crédito)
df['BUREAU_CREDIT_AGE_MEAN'] = df['BUREAU_DAYS_CREDIT_MEAN'].abs()  # Tempo médio desde abertura dos créditos
```

```python
''' Feature: crédito mais recente no bureau
Importância: histórico muito recente está associado a maior risco '''

# Converter o crédito mais recente do bureau para valor positivo
df['BUREAU_CREDIT_RECENT'] = df['BUREAU_DAYS_CREDIT_MAX'].abs()  # Tempo desde o crédito mais recente
```

```python
''' Feature: indicador de histórico curto no bureau
Importância: clientes novos no crédito têm risco  '''

# Criar flag para identificar histórico de crédito curto no bureau
df['SHORT_CREDIT_HISTORY'] = (
    df['BUREAU_CREDIT_AGE_MEAN'] < 365  # Histórico médio menor que 1 ano
).astype(int)                          # 1 = histórico curto
```

```python
''' Feature: atraso médio nos contratos POS
Importância: histórico de atraso indica comportamento de risco '''

# Aplicar transformação logarítmica no atraso médio (DPD) para reduzir assimetria
df['POS_DPD_MEAN_LOG'] = np.log1p(df['POS_SK_DPD_MEAN'])  # log(1 + dias médios em atraso)
```

```python
''' Feature: indicador de atraso relevante
Importância: clientes com DPD apresentam maior inadimplência '''

# Criar flag para indicar se o cliente já teve atraso em pagamentos (DPD > 0)
df['HAS_DPD_FLAG'] = (
    df['POS_SK_DPD_MEAN'] > 0  # Qualquer atraso registrado
).astype(int)                 # 1 = já teve atraso
```

```python
''' Feature: dívida ativa média no bureau
Importância: mede exposição média a dívidas ativas '''

# Aplicar transformação logarítmica na média da dívida ativa no bureau
df['BUREAU_ACTIVE_DEBT_MEAN_LOG'] = np.log1p(
    df['BUREAU_AMT_CREDIT_SUM_DEBT_MEAN']  # log(1 + dívida média ativa)
)
```

```python
''' Feature: proporção de dívida ativa no total de crédito
Importância: indica quanto do crédito contratado ainda está aberto '''

# Calcular proporção da dívida ativa em relação ao crédito total no bureau
df['BUREAU_DEBT_RATIO'] = (
    df['BUREAU_AMT_CREDIT_SUM_DEBT_SUM'] /      # Dívida ativa total
    (df['BUREAU_AMT_CREDIT_SUM_SUM'] + 1)       # Crédito total (+1 evita divisão por zero)
)
```

```python
''' Feature: indicador de alta alavancagem no bureau
Importância: clientes muito alavancados concentram maior risco '''

# Criar flag para identificar alta alavancagem no bureau de crédito
df['HIGH_BUREAU_LEVERAGE'] = (
    df['BUREAU_DEBT_RATIO'] > 0.50  # Mais de 50% do crédito comprometido com dívida
).astype(int)                      # 1 = alta alavancagem
```

```python
''' Feature: razão entre valor pago e valor da parcela
Importância: mede se o cliente paga integralmente suas obrigações '''

# Calcular a razão média entre valor pago e valor da parcela
df['PAYMENT_RATIO_MEAN'] = (
    df['INST_AMT_PAYMENT_MEAN'] /
    (df['INST_AMT_INSTALMENT_MEAN'] + 1)  # +1 evita divisão por zero
)
```

```python
''' Feature: taxa de propostas recusadas
Importância: múltiplas recusas indicam alto risco percebido '''

# Calcular taxa de recusas em créditos anteriores
df['PREV_REFUSAL_RATE'] = (
    df['PREV_STATUS_REFUSED'] /                         # Número de propostas recusadas
    (df['PREV_STATUS_REFUSED'] +                         # Recusadas
     df['PREV_STATUS_APPROVED'] + 1)                     # Aprovadas (+1 evita divisão por zero)
)
```
